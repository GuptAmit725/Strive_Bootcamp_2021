Do the same thing as you did for the previous exercise, but now process the text line by line. This is something that you would have to do if you had to process a very long text (as it is the case).

```python
# Counting words line by line.
fp = open(r"C:\Users\amiti\Desktop\Strive\Strive_Bootcamp_2021\M1\W1\D4\d4_Strings_and_files\text-files\blakepoems.txt",'r+')
words = []
for i in fp.readlines(): #  reading lineby line
    i = i.replace('[','')
    i = i.replace(']','')
    i = i.replace('\n','')
    if i.strip():
        for w in i.split(' '): # fetching all the words which is not a letter
            if len(w)>1: 
                words.append(w.lower())
        
# coumting all the words
count_dict = {}
for  word in words:
    count_dict[word] = words.count(word)

count_dict
```
